# Сбор датасета

Шлем запросы в inaturalist.com с фильтрами на определенный таксон и тип лицензии. При этом анализируем фотографии несколькими функциями CV, чтобы сразу заранее понять, какие из отобранных фотографий имеют плохое качество и могут вызвать трудности при разметке. Такие кейсы тоже будут исключительно полезны.

В процессе скачивания данных становится ясно, что inaturalist, как тоже своего рода краудсорсинговая платформа, обременена классическими проблемами: люди неверно определяют бабочек, шлют плохие фотографии, среди фотографий бабочек замешаны фотографии гусениц. 

Хорошо, что мы осведомлены об этих проблемах еще на этапе сбора датасета, так как это позволит нам составить более качественную инструкцию.

```python
flower_target_photo_count = 200
flower_taxon_id = 47125  

flower_df = collect_inat_photos(
    taxon_id=flower_taxon_id,
    target_photo_count=flower_target_photo_count,
    photo_license="cc0",
    unique_authors=False,
    annotate_hard=True,
    add_info=False,
)
flower_df.head(3)

save_df_json(flower_df, "../data/flowers.json")
```
